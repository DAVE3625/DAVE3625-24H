```python
#stud.isnull()

#Lets first replace all empty cells wit np.nan
stud=stud.replace(r'^\s*$', np.nan, regex=True)
#and check again
stud.isna().sum()

#stud=stud.replace(r'^\s*$', np.nan, regex=True) #Replace blank values with np.nan values

#stud=stud.replace(r'^\s*$', np.nan, regex=True) #Replace blank values with np.nan values
```

```python
#When dealing with missing values, we need to make to figure out if we can replace data
#with dummy values. Lets say Age isn't too important in this dataset, and we can replace missing
#age values with 0

stud"Age"].replace(np.nan, 0, inplace=True)
#df = df['Age'].astype(str).astype(int)
```

```python
stud.isna().sum()
```

```python
#Since we believe the rest of the columns contains important data, we won't replace nan values with dummy data
#We will therefor delete rows containing nan data
stud.dropna(inplace = True)
# inplace = True updates the df with the new data, alternative you can do df = df.dropna() or Newdf = df.dropna()
#And lets check again
stud.isna().sum()
```

```python
stud['Age'] = stud['Age'].fillna(0)
stud['Age'] = stud['Age'].astype(str).astype(int) #Convert from obj to int

stud['hrsStudy'] = stud['hrsStudy'].fillna(0)
stud['hrsStudy'] = stud['hrsStudy'].astype(str).astype(int) #Convert from obj to int

stud.info()
stud.isna().sum()

#stud['Age'] = pd.to_numeric(stud['Age'], errors='coerce')
#stud['hrsStudy'] = pd.to_numeric(stud['hrsStudy'], errors='coerce')
```

```python
#As you can see, the index counts to 49, but we have dropped several values.. to reset the index you can use
stud= stud.reset_index(drop=True)
stud.tail(5)
```

```python
# create a list of our conditions
conditions = [
    (stud['FinalGrade'] <= 50.0),
    (stud['FinalGrade'] > 50.0) and (stud['FinalGrade'] <= 60.0),
    (stud['FinalGrade'] > 60.0) and (stud['FinalGrade'] <= 70.0),
    (stud['FinalGrade'] > 70.0) and (stud['FinalGrade'] <= 80.0),
    (stud['FinalGrade'] > 80.0) and (stud['FinalGrade'] <= 90.0),
    (stud['FinalGrade'] > 90.0)
    ]

# create a list of the values we want to assign for each condition
values = ['F', 'E', 'D', 'C', 'B', 'A']

# create a new column and use np.select to assign values to it using our lists as arguments
stud['Grade'] = np.select(conditions, values)

# display updated DataFrame
stud.head()


#Grades_new_col = [
#    (stud['FinalGrade'] <= 50.0),
#    (stud['FinalGrade'] > 50.0) and(stud['FinalGrade'] <= 60.0),
#    (stud['FinalGrade'] > 60.0) and (stud['FinalGrade'] <= 70.0),
#    (stud['FinalGrade'] > 70.0) and (stud['FinalGrade'] <= 80.0),
#    (stud['FinalGrade'] > 80.0) and (stud['FinalGrade'] <= 90.0),
#    (stud['FinalGrade'] > 90.0)
#    ]


#Letter_grades = ['F', 'E', 'D', 'C', 'B', 'A']

#stud['Grade'] = np.select(Grades_new_col,Letter_grades)
#stud = pd.DataFrame(Grades_new_col) # Create a DataFrame from the data
#stud.head()

```

```python
#We can now see how many got the different grades by using groupby
stud_gradeCount = stud.groupby('Grade').count()
stud_gradeCount.FinalGrade
```
